```python
# Task 1: Exploratory Data Analysis (EDA) and Business Insights

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
customers = pd.read_csv(r"Customers.csv")
products = pd.read_csv(r"Products.csv")
transactions = pd.read_csv(r"Transactions.csv")
```

```python
# Display basic information
print("Customers Dataset:\n", customers.info(), "\n")
print("Products Dataset:\n", products.info(), "\n")
print("Transactions Dataset:\n", transactions.info(), "\n")

# Display the first few rows
print("\nFirst few rows of Customers:\n", customers.head())
print("\nFirst few rows of Products:\n", products.head())
print("\nFirst few rows of Transactions:\n", transactions.head())
```

```python
# Check for missing values
print("\nMissing Values in Customers:\n", customers.isnull().sum())
print("\nMissing Values in Products:\n", products.isnull().sum())
print("\nMissing Values in Transactions:\n", transactions.isnull().sum())

# Basic statistics
print("\nBasic Statistics for Transactions:\n", transactions.describe())
```

```python
# Customers: Analyze region distribution
region_counts = customers['Region'].value_counts()
plt.figure(figsize=(8, 5))
region_counts.plot(kind='bar', color='skyblue')
plt.title("Customer Distribution by Region")
plt.xlabel("Region")
plt.ylabel("Number of Customers")
plt.show()

# Customers: Analyze signup trends
customers['SignupDate'] = pd.to_datetime(customers['SignupDate'], errors='coerce')
date_counts = customers['SignupDate'].dt.year.value_counts().sort_index()
plt.figure(figsize=(8, 5))
date_counts.plot(kind='bar', color='orange')
plt.title("Customer Signups by Year")
plt.xlabel("Year")
plt.ylabel("Number of Signups")
plt.show()

# Products: Analyze categories
category_counts = products['Category'].value_counts()
plt.figure(figsize=(8, 5))
category_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title("Product Distribution by Category")
plt.ylabel("")
plt.show()

# Transactions: Analyze transaction trends
transactions['TransactionDate'] = pd.to_datetime(transactions['TransactionDate'], errors='coerce')
date_transaction = transactions['TransactionDate'].dt.year.value_counts().sort_index()
plt.figure(figsize=(8, 5))
date_transaction.plot(kind='line', marker='o', color='green')
plt.title("Transactions Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Transactions")
plt.show()

# Transactions: Analyze quantity vs total value
plt.figure(figsize=(8, 5))
sns.scatterplot(data=transactions, x='Quantity', y='TotalValue', hue='Price', palette='viridis')
plt.title("Quantity vs Total Value by Price")
plt.xlabel("Quantity")
plt.ylabel("Total Value")
plt.show()
```

```python
# Save clean data for further use
customers.to_csv("Clean_Customers.csv", index=False)
products.to_csv("Clean_Products.csv", index=False)
transactions.to_csv("Clean_Transactions.csv", index=False)
```
